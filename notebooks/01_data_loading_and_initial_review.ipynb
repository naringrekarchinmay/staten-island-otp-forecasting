# ==============================
# DATA CLEANING
# ==============================

```python
df.duplicated().sum()
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_percentage
```

```python
df[df['Incomplete Trips'].isnull()].head()
```

```python
df[df['Trip Complete Percentage'].isnull()].head()
```

```python
df['Year'] = df['Month'].dt.year
df['Month_Number'] = df['Month'].dt.month
df['Month_Name'] = df['Month'].dt.month_name()
```

```python
df.head()
```

```python
df['Day Time'].unique()
```

```python
df['Trip Complete Percentage'].unique()
```

```python
df['Trip Complete Percentage'] = (
    df['Trip Complete Percentage']
    .astype(str)
    .str.replace('%', '', regex=False)
)

df['Trip Complete Percentage'] = pd.to_numeric(
    df['Trip Complete Percentage'],
    errors='coerce'
)
```

```python
df.info()
```

```python
df.to_csv('../data/raw/cleaned_staten_island_otp.csv', index=False)
```
