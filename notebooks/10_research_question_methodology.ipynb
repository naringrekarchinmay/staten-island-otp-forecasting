# Phase 16: Research Question and Methodology

## Staten Island Railway OTP Forecasting Project

This notebook formally defines the research question, supporting research questions, hypotheses, methodology, evaluation strategy, and expected contribution of the Staten Island Railway On-Time Performance forecasting project.

The purpose of this phase is to move the project from a technical machine learning implementation into a structured research study suitable for a graduate-level final project, academic presentation, or potential publication draft.

## 1. Project Context

On-time performance is one of the most important indicators of transit reliability. For commuter rail systems, even small changes in OTP can affect passenger experience, operational planning, scheduling decisions, and service quality.

This project focuses on monthly Staten Island Railway on-time performance forecasting using historical MTA open data. Earlier phases of the project focused on data cleaning, exploratory data analysis, feature engineering, machine learning model development, explainability, baseline model comparison, time-aware cross-validation, and prediction intervals.

The current phase formally defines the research question and methodology so that the project can be presented as a rigorous applied machine learning study rather than only a predictive modeling exercise.

## 2. Research Objective

The objective of this project is to evaluate whether machine learning models can improve short-horizon on-time performance forecasting for the Staten Island Railway when compared with traditional statistical time-series forecasting methods.

The project also aims to determine whether model explainability techniques can provide useful operational insight into the factors influencing predicted OTP.

The broader goal is to support proactive transit operations planning by forecasting future OTP trends, identifying potential risk periods, and explaining the historical patterns that influence model predictions.

## 3. Main Research Question

Can gradient-boosted tree models using lag-based, rolling-window, calendar, and operational features outperform traditional statistical time-series baselines such as SARIMA and Prophet in forecasting short-horizon commuter rail on-time performance, while providing interpretable operational insights through SHAP analysis?

## 4. Supporting Research Questions

### RQ1: Forecasting Accuracy

Do machine learning models produce lower prediction error than traditional statistical forecasting models for monthly Staten Island Railway OTP forecasting?

### RQ2: Feature Importance and Interpretability

Which historical, seasonal, and operational features contribute most strongly to OTP prediction?

### RQ3: Time-Aware Validation

How stable is model performance when evaluated using time-aware cross-validation instead of a single train-test split?

### RQ4: Forecast Uncertainty

Can prediction intervals provide useful uncertainty estimates around OTP forecasts for operational decision-making?

### RQ5: Operational Usefulness

Can the forecasting framework support proactive planning by identifying future months or periods with increased risk of lower OTP?

## 5. Research Hypotheses

### H1: Model Performance

Machine learning models, particularly XGBoost, will achieve lower MAE and RMSE than SARIMA and Prophet baselines for short-horizon OTP forecasting.

### H2: Historical Dependency

Lag-based OTP features and rolling-window averages will be among the strongest predictors of future OTP.

### H3: Validation Rigor

Time-aware cross-validation will provide a more realistic estimate of forecasting performance than a single train-test split.

### H4: Explainability

SHAP explainability will reveal interpretable relationships between historical OTP behavior, seasonal patterns, and predicted OTP.

### H5: Decision Support

Prediction intervals and future risk classification can make the forecasting framework more useful for operational planning than point forecasts alone.

## 9. Feature Engineering

The forecasting problem was transformed into a supervised machine learning problem through feature engineering.

The engineered features included:

- Lag-based OTP features
- Rolling-window averages
- Calendar features
- Trend indicators
- Seasonal indicators

Examples of lag features include previous-month OTP and OTP values from earlier months. Rolling-window features summarize recent historical performance over multiple months.

These features allow machine learning models to learn from historical performance behavior and temporal structure.

### Statistical Baseline Models

The baseline models include:

- Prophet

### Machine Learning Models

## 11. Validation Strategy

Because this is a time-series forecasting problem, random train-test splitting is not appropriate.

Future observations should not be used to predict past observations. To avoid data leakage and preserve chronological order, this project uses time-aware validation.

The main validation method is:

**TimeSeriesSplit cross-validation**

This approach evaluates the model across multiple expanding time windows. Each fold trains on earlier observations and tests on later observations.

This provides a more realistic estimate of how the model may perform when forecasting future OTP values.

## 12. Evaluation Metrics

Model performance is evaluated using the following metrics:

### Mean Absolute Error

MAE measures the average absolute difference between actual OTP and predicted OTP.

In this project, MAE is especially useful because it can be interpreted directly in OTP percentage points.

For example, an MAE of 2.11 means the model is off by approximately 2.11 OTP percentage points on average.

### Root Mean Squared Error

RMSE measures forecast error while penalizing larger errors more heavily. This is useful for identifying models that may occasionally produce large prediction mistakes.

### R-Squared

R² measures how much variation in OTP is explained by the model. While useful, it should be interpreted carefully because time-series datasets with limited observations can produce unstable R² values.

## 13. Explainability Methodology

SHAP analysis is used to interpret the best-performing machine learning model.

The purpose of SHAP is to explain how individual features contribute to model predictions.

In this project, SHAP helps answer the following questions:

- Which features are most influential in predicting OTP?
- Do recent OTP values strongly influence future OTP?
- Do seasonal or calendar features contribute meaningfully to prediction?
- Are the model's predictions operationally interpretable?

This is important because transit forecasting models should not only produce accurate predictions but also provide insight that can support decision-making.

## 14. Forecast Uncertainty

Point forecasts provide a single predicted OTP value. However, operational planning often requires understanding uncertainty.

To address this, prediction intervals were added to the forecasting workflow.

Prediction intervals provide a likely range of future OTP values instead of only one estimate.

This allows the forecast to communicate both the expected OTP and the uncertainty around that prediction.

For transit operations, this is useful because a forecast with high uncertainty may require more cautious interpretation than a forecast with a narrow prediction interval.

## 15. Methodology Summary

The complete methodology follows these steps:

1. Collect monthly Staten Island Railway OTP data from MTA open data.
2. Clean and prepare the data for time-series analysis.
3. Conduct exploratory data analysis to identify trends, seasonality, and variation.
4. Transform the time-series forecasting problem into a supervised learning problem using lag, rolling-window, calendar, and trend features.
5. Train traditional statistical baseline models including SARIMA and Prophet.
6. Train machine learning models including Linear Regression, Random Forest, and XGBoost.
7. Evaluate models using MAE, RMSE, and R².
8. Use TimeSeriesSplit cross-validation to test model stability over multiple chronological folds.
9. Apply SHAP explainability to interpret the best-performing machine learning model.
10. Add prediction intervals to communicate forecast uncertainty.
11. Use the results to support future OTP risk classification and operational decision-support analysis.

## 16. Expected Contribution

This project contributes to applied transit analytics in three ways.

First, it compares machine learning models against traditional statistical time-series baselines for commuter rail OTP forecasting.

Second, it combines forecasting accuracy with explainability by using SHAP to identify the factors that influence predicted OTP.

Third, it introduces uncertainty estimation through prediction intervals, making the forecasting output more useful for operational decision-making.

Together, these components create a forecasting framework that is more rigorous than a simple point-prediction model and more practical for transit performance analysis.

## 17. Limitations

This project has several limitations.

First, the dataset is based on monthly observations, which limits the number of available training samples. A larger dataset with daily or trip-level observations may allow for more detailed forecasting.

Second, the current version primarily relies on historical OTP patterns. Additional external variables such as weather, incidents, ridership, maintenance activity, special events, and staffing levels may improve model performance.

Third, the model is currently focused on Staten Island Railway. Testing the framework on another transit dataset would help evaluate whether the methodology generalizes to other systems.

Fourth, prediction intervals provide useful uncertainty estimates, but they should be interpreted as approximate ranges rather than guaranteed outcomes.

Finally, the project is designed as a decision-support tool and should not replace expert operational judgment.
